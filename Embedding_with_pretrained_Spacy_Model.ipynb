```python
nlp = spacy.load("en_core_web_md")
```

```python
doc = nlp("I deposited money in the bank")
```

```python
doc
```

```python
for token in doc:
  print(token.text, token.has_vector, token.vector_norm, token.is_oov)
```

```python
for token in doc:
    print(token.vector)
```
