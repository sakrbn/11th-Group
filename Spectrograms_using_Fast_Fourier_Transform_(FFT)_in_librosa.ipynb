```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
```

```python
audio_filepath = "/content/NLP_sample_audio.wav"
```

```python
y , sr = librosa.load(audio_filepath)
```

```python
S = librosa.stft(y)
spectrogram = np.abs(S)
```

```python
spectrogram
```

```python
plt.figure(figsize=(12, 6))  # Adjust figure size as desired
plt.imshow(spectrogram, aspect='auto', origin='lower', cmap='inferno')  # Choose a desired colormap
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.title('Spectrogram of the Audio')
plt.colorbar(format='%2.f dB')  # Display colorbar with dB scale
plt.show()
```

```python
S_dB = librosa.amplitude_to_db(np.abs(S), ref=np.max)
```

```python
plt.figure(figsize=(10, 5))
librosa.display.specshow(S_dB, sr=sr, x_axis='time', y_axis='log', cmap='inferno')
plt.colorbar(format='%+2.0f dB')
plt.title('STFT Spectrogram')
plt.xlabel('Time')
plt.ylabel('Frequency (Hz)')
plt.show()
```
