```python
import numpy as np
import librosa
import matplotlib.pyplot as plt
```

```python
y, sr = librosa.load('/content/NLP_sample_audio.wav')
```

```python
order = 16
lpc_coefficients = librosa.lpc(y, order=order)
```

```python
lpc_coefficients
```

```python
lpc_coefficients.shape
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(2,1,1)
plt.plot(lpc_coefficients.T, label='LPC Coefficients')
plt.xlabel('Frame Index')
plt.ylabel('LPC Coefficients')
plt.title(f'LPC Coefficients(order={order})')
plt.legend()
```
